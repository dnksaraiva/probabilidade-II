#### Primeiramente, vamos calcular o valor de C.

Temos que,

$$c \int\limits_{-1}^{1} 1 - x^2 \, dx$$

A propriedade da Integral diz

 $$\int a  \bullet f(x)dx = a  \int  f(x) \,dx \, a  \, \epsilon \, \,   \mathbb{R}$$
 
Assim, primeiramente vamos calcular a integral indefinida

$$c \int 1 - x^2 \, dx$$


$$c  \bullet  \bigg( \int 1 \,dx -  \int x^2 \,dx  \bigg)$$

$$c  \bullet  \bigg(x -  \frac{x^3}{3} \bigg)$$

$$cx - \frac{cx^3}{3}$$


Agora, para calcular a integral definida, adiciona-se os limites de integração.

    Pela propriedade, temos que:
    
$$F(x) \, \bigg|_a^b = F(b) - F(a)$$    

    Logo, 
    
$$\bigg(c x - \frac{cx^3}{3} \bigg)\bigg|_{-1}^{1}$$
    
    Para F(x) = F(1) - F(-1)
    
$$\bigg(c . 1 - \frac{c . 1^3}{3}\bigg) \, - \, \bigg(c . (-1) \frac {c . (-1)^3}{3}\bigg)$$


$$\bigg(c - \frac{c}{3}\bigg) \, - \, \bigg(-c +\frac{c}{3}\bigg)$$

$$\bigg(c - \frac{c}{3}\bigg) \, - \, \bigg(-\frac{2}{3} .c\bigg)$$

$$c -\frac{c}{3} \, + \, \frac{2}{3}$$

$$\frac{4}{3} . c$$

$$c = \frac{3}{4}$$


Calculando a Esperança de X, pela definição 

$$\mu = E(x) =  \int_{-\infty}^{\infty} f(x)\, dx$$

$$\int_{-1}^{1} x \, \frac{3}{4} \, (1 - x^2) \,dx$$


Calculando a integral indefinida 

$$\int x \frac{3}{4} . (1-x^2)\,dx$$

    propriedade
    
$$\int a f(x)\,dx = a\int f(x)\,dx, a \epsilon \, \,   \mathbb{R}$$

$$\frac{3}{4} \int x(1-x^2) \,dx$$

$$\frac{3}{4} \int x - x^3 \,dx$$

    propriedade 

$$\int f(x) \pm g(x)\,dx = \int f(x)\,dx \pm \int g(x)\,dx$$

$$\frac{3}{4} \bigg(\int x dx - \int x^3dx\bigg)$$

$$\frac{3}{4} \bigg(\frac{x^2}{2} - \frac{x^4}{4}\bigg)$$

$$\frac{3x^2}{8} - \frac{3x^4}{16}$$

Adicionando os limites de integração pela propriedade

$$F(x) \, \bigg|_a^b = F(b) - F(a)$$  

$$\bigg(\frac {3x^2}{8} - \frac {3x^4}{16}\bigg) \bigg|_{-1}^{1}$$

$$\bigg(\frac {3 . 1^2}{8} - \frac {3 . 1^2}{16}\bigg) - \bigg(\frac{3.(-1^2)}{8} - \frac{3 . (-1^4)}{16}\bigg) = 0$$


Calculando a variância de X.

$$E(x^2) = \int_{-\infty}^{\infty} x^2f(x)\,dx$$

$$\int_{-1}^2{1} x^2 \frac{3}{4} (1-x^2)\,dx$$

Calculando a integral indefinida

$$\int x^2 \frac{3}{4} (1-x^2) \,dx$$

$$\frac{3}{4}\int x^2(1-x^2)\,dx$$

$$\frac{3}{4}\int x^2 - x^4 \,dx$$

$$\frac{3}{4} \bigg(\int x^2\,dx - \int x^4\,dx\bigg)$$

$$\frac{3}{4}\bigg(\frac{x^3}{3} \,-\,\frac{x^5}{5} \bigg)$$

$$\frac{3x^3}{12} - \frac{3x^5}{20}$$

    Temos, 
    
$$F(x) \, \bigg|_a^b = F(b) - F(a)$$

    Logo,
    
$$\frac{3x^3}{12} - \frac{3x^5}{20} \, \, \,\bigg|_{-1}^{1}$$

$$\bigg(\frac{3 . 1^3}{12} - \frac{3 . 1^5}{20}\bigg) - \bigg(\frac{3 . (-1^3)}{12} - \frac{3 . (-1^5)}{20}\bigg) = \frac{1}{5}$$

O Coeficiente de Assimetria é calculado como 

$$\alpha_3 \, = \, \frac{\mu_3}{\sigma_3}$$

em que $\mu_3$ é o terceiro momento central, sabemos que 

$$\mu_3 \, = E(X - \mu)^3 = E(X)^3 - 3E(X^2)E(X)+2E(X)^3$$

Como a distribuição é simétrica o Coeficiente de Assimetria é nulo.

    Provando
    
$$E(X^3) = \int_{-\infty}^{\infty} \, x^3 f(x)\,dx$$
    
    Logo, 
    
$$\int_{-1}^{1}\,x^3\frac{3}{4}\,(1-x^2)\,dx$$

    Calculando a integral indefinida

$$\int x^3 \frac{3}{4}(1-x^2)dx$$

$$\frac{3}{4}\int x^3 (1-x^2)dx$$

$$\frac{3}{4}\int x^3 \, - \, x^5dx$$

$$\frac{3}{4}\bigg(\int x^3dx \, - \, \int x^5dx \bigg)$$

$$\frac{3}{4}\bigg(\frac{x^4}{4} \, - \,\frac{x^6}{6}\bigg)$$

$$\frac{3x^4}{16}\,-\,\frac{x^6}{8}$$

    Calculando a integral definida
    
$$\bigg(\frac{3x^4}{16}\,-\,\frac{x^6}{8}\bigg)\,\,\, \bigg|_{-1}^{1}$$

$$\bigg(\frac{3 . 1^4}{16} \,-\, \frac{1^6}{8}\bigg) \,-\, \bigg(\frac{3(-1)^4}{16}\,-\,\frac{(-1)^6}{8}\bigg) = 0$$

    Assim, 
    
$$\mu_3 = E(X^3) - 3E(X^2)E(X)+2E(X)^3$$

$$0 - \bigg(3 \frac{1}{5}\bigg) .0 + (2 . 0^3) = 0$$

Logo, $$\alpha_3 = 0$$

O coeficiente de assimetria é adimensional.


O Coeficiente de Curtose é calculado através de:

$$\alpha_4 \, = \, \frac{\mu_4}{\sigma_4}$$

em que $\mu_4$ é o quarto momento central.

Sabemos que

$$\mu = E (X − \mu)^4 = E(X^4) − 4E(X^3)E(X) + 6E(X^2)E^2(X) − 3E^4(X)$$

Vamos calcular o quarto momento em relação à origem.
    
    Provando

$$E(X^4) = \int_{-\infty}^{\infty} \, x^4 f(x)\,dx$$

    Logo, 
    
$$\int_{-1}^{1}\,x^4\frac{3}{4}\,(1-x^2)\,dx$$

    Calculando a integral indefinida

$$\int x^4 \frac{3}{4} (1-x^2)\,dx$$

$$\frac{3}{4}\int x^4(1-x^2)\,dx$$

$$\frac{3}{4}\int x^4 - x^6 \,dx$$

$${3}{4}\bigg(\int x^4dx \, - \,\int x^6dx \bigg)$$

$$\frac{3}{4}\bigg(\frac{x^5}{5} \,-\, \frac{x^7}{7}\bigg)$$

$$\frac{3x^5}{20}\,-\, \frac{3x^7}{28}$$

Calculando a integral definida, para $F(x)\,\bigg|_a^b = F(b) - F(a)$

$$\bigg(\frac{3x^5}{20}\,-\,\frac{3x^7}{28}\bigg)\bigg|_{-1}^{1}$$

$$\bigg(\frac{3 . 1^5}{20}\,-\,\frac{3 . 1^7}{28}\bigg)\,-\,\bigg(\frac{3(-1)^5}{20}\,-\,\frac{3(-1)^7}{28}\bigg) = \frac{3}{35}$$

Vamos calcular o quarto momento central. 

$$\mu_4 = E(X^4) − 4E(X^3)E(X) + 6E(X^2)E^2(X) − 3E^4(X)$$

$$\mu_4 = \frac{3}{35} - 4 . 0 . 0 + 6 \frac{1}{5} . 0^2 - 3 . 0^4$$

$$\mu_4 = \frac{3}{35}$$

O coeficiente de curtose é dado por:

$$\alpha_4 = \frac{\mu_4}{\sigma_4} = \frac{\mu_4}{V^2(X)}$$

$$\alpha_4 = \frac{\frac{3}{35}}{\frac{1^5}{5^2}} = \frac{3}{35} . \frac{5^2}{1^2}$$

$$\frac{75}{35} ≈ 2,14 < 3$$

Logo, a distribuição é platicúrtica.
